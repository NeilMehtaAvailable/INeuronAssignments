```python
# Question1. Create a function that takes three arguments a, b, c and returns the sum of the
# numbers that are evenly divided by c from the range a, b inclusive.
# Examples
# evenly_divisible(1, 10, 20) ➞ 0
# # No number between 1 and 10 can be evenly divided by 20.
# evenly_divisible(1, 10, 2) ➞ 30
# # 2 + 4 + 6 + 8 + 10 = 30
# evenly_divisible(1, 10, 3) ➞ 18
# # 3 + 6 + 9 = 18
#
def multiinput(listofnums):
    a, b, c = int(listofnums[0]), int(listofnums[1]), int(listofnums[2])
    summing = []
    for j in range(a, b + 1):
        if j % c == 0:
            summing.append(j)
    print("The sum of divisble numbers is:", sum(summing))


inputlist = []
for i in range(0,3):
    inputlist.append(int(input("Enter numbers: ")))

for l in range(0,3):
    multiinput(inputlist)
```

```python
# Question2. Create a function that returns True if a given inequality expression is correct and
# False otherwise.
# Examples
# correct_signs("3 < 7 < 11") ➞ True
# correct_signs("13 > 44 > 33 > 1") ➞ False
# correct_signs("1 < 2 < 6 < 9 > 3") ➞ True
#
def correct_signs(inputstring):
    inputlist = inputstring.split(" ")
    numberlist = []
    symbollist = []
    collectionlist = []


    for valandchar in inputlist:
        if valandchar.isnumeric():
            numberlist.append(int(valandchar))
        elif valandchar == "<" or valandchar == ">" or valandchar == "==":
            symbollist.append(valandchar)
        else:
            return print("Invalid expression. Enter  expression containing <,>,= seperated with numbers by a space")


    for i in range(0, int(len(symbollist))):
        if symbollist[i] == "<":
            collectionlist.append(numberlist[i] < numberlist[i + 1])
        elif symbollist[i] == ">":
            collectionlist.append(numberlist[i] > numberlist[i + 1])
        elif symbollist[i] == "==":
            collectionlist.append(numberlist[i] == numberlist[i + 1])


    return print(False not in collectionlist)
    return print(False in collectionlist)


inputexpression = input("Enter expression where numbers and signs are seperated by white spaces: ")
correct_signs(inputexpression)
```

```python
# Question4. Write a function that calculates the factorial of a number recursively.
# Examples
# factorial(5) ➞ 120
# factorial(3) ➞ 6
# factorial(1) ➞ 1
# factorial(0) ➞ 1
#
def factorialfunc(x, previousnum):
    if x > 1:
        previousnum = previousnum * x
        y = x - 1
        factorialfunc(y, previousnum)
    elif x == 1:
        return print(previousnum)
    else:
        return print("Positive integers only")



factorialinput = int(input("Enter the number whose factorial  you want to find: "))
factorialfunc(factorialinput, 1)
```

```python
# Question 5
# Hamming distance is the number of characters that differ between two strings.
# To illustrate:
# String1: "abcbba"
# String2: "abcbda"
# Hamming Distance: 1 - "b" vs. "d" is the only difference.
# Create a function that computes the hamming distance between two strings.
# Examples
# hamming_distance("abcde", "bcdef") ➞ 5
# hamming_distance("abcde", "abcde") ➞ 0
# hamming_distance("strong", "strung") ➞ 1
def hammingnumber(inputstr1, inputstr2):
    counter = 0
    if int(len(inputstr1)) == int(len(inputstr2)):
        inputlist1 = list(inputstr1)
        inputlist2 = list(inputstr2)
        for i in range(0, int(len(inputlist1))):
            if inputlist1[i] != inputlist2[i]:
                counter += 1
        return print(counter)
    else:
        return print("Enter strings of equal length")


input1 = input("Enter first string: ")
input2 = input("Enter second string: ")
hammingnumber(input1,input2)
```
