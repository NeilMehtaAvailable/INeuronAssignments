```python
# Question1
# Create a function that takes a string and returns a string in which each character is repeated
# once.
# Examples
# double_char("String") ➞ "SSttrriinngg"
# double_char("Hello World!") ➞ "HHeelllloo WWoorrlldd!!"
# double_char("1234!_ ") ➞ "11223344!!__ "
#
def double_char(inputstring):
    outputstring = ""
    for i in inputstring:
        outputstring = outputstring + i*2


    return print(outputstring)


requestedstring = input("Enter a word or sentence you want to ee alphabets doubled: ")
double_char(requestedstring)
```

```python
# Question2
# Create a function that reverses a boolean value and returns the string "boolean expected"
# if another variable type is given.
# Examples
# reverse(True) ➞ False
# reverse(False) ➞ True
# reverse(0) ➞ "boolean expected"
# reverse(None) ➞ "boolean expected"
#
def reverse(inputvalue):
    if inputvalue == "True":
        outputvalue = False
    elif inputvalue == "False":
        outputvalue = True
    else:
        outputvalue = "boolean expected"
    return print(outputvalue)


inputval = input("Enter the value for boolean reversal: ")
reverse(inputval)
```

```python
# Question3
# Create a function that returns the thickness (in meters) of a piece of paper after folding it n
# number of times. The paper starts off with a thickness of 0.5mm.
# Examples
# num_layers(1) ➞ "0.001m"
# # Paper folded once is 1mm (equal to 0.001m)
# num_layers(4) ➞ "0.008m"
# # Paper folded 4 times is 8mm (equal to 0.008m)
# num_layers(21) ➞ "1048.576m"
# # Paper folded 21 times is 1048576mm (equal to 1048.576m)
#
def num_layers(inputfolds):
    outputthicknessinmm = (2 ** int(inputfolds)) * 0.5
    outputthicknessinm = outputthicknessinmm / 1000
    if int(inputfolds) == 1:
        return print("Paper folded once is", int(outputthicknessinmm), "\bmm", "(equal to", outputthicknessinm, "\bm)")
    elif int(inputfolds) == 2:
        return print("Paper folded twice is", int(outputthicknessinmm), "\bmm", "(equal to", outputthicknessinm, "\bm)")
    elif int(inputfolds) == 3:
        return print("Paper folded thrice is", int(outputthicknessinmm), "\bmm", "(equal to", outputthicknessinm, "\bm)")
    else:
        return print("Paper folded", inputfolds,"times is", int(outputthicknessinmm), "\bmm", "(equal to", outputthicknessinm, "\bm)")


inputvalue = input("Enter the number of times you want a 0.5mm paper you want to fold: ")
num_layers(inputvalue)
```

```python
# Question4
#
# Create a function that takes a single string as argument and returns an ordered list containing
# the indices of all capital letters in the string.
# Examples
# index_of_caps("eDaBiT") ➞ [1, 3, 5]
# index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]
# index_of_caps("determine") ➞ []
# index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]
# index_of_caps("sUn") ➞ [1]
#
def index_of_caps(inputstring):
    resultlist = []
    for i in range(0, int(len(inputstring))):
        if inputstring[i].isupper():
            resultlist.append(i)


    return print(resultlist)


inputvalue = input("Enter a word or sentence with mixed case: ")
index_of_caps(inputvalue)
```

```python
# Question5
# Using list comprehensions, create a function that finds all even numbers from 1 to the given
# number.
# Examples
# find_even_nums(8) ➞ [2, 4, 6, 8]
# find_even_nums(4) ➞ [2, 4]
# find_even_nums(2) ➞ [2]
def inputlimit(upperlimit):
    return print([i for i in range(0, int(upperlimit) + 1) if i % 2 == 0])


inputlimit = input("Enter the upper limit to find alleven numbers: ")
inputlimit(inputlimit)
```
