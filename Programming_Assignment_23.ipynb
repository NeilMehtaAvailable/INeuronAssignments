```python
# Question 1
# Create a function that takes a number as an argument and returns True or False depending
# on whether the number is symmetrical or not. A number is symmetrical when it is the same as
# its reverse.
# Examples
# is_symmetrical(7227) ➞ True
# is_symmetrical(12567) ➞ False
# is_symmetrical(44444444) ➞ True
# is_symmetrical(9939) ➞ False
# is_symmetrical(1112111) ➞ True
#
def is_symmetrical(inputnumber):
    return print(inputnumber == inputnumber[::-1])


inputnum = input("Enter the number to check symmetricity: ")
is_symmetrical(inputnum)
```

```python
# Question 2
# Given a string of numbers separated by a comma and space, return the product of the
# numbers.
# Examples
# multiply_nums("2, 3") ➞ 6
# multiply_nums("1, 2, 3, 4") ➞ 24
# multiply_nums("54, 75, 453, 0") ➞ 0
# multiply_nums("10, -2") ➞ -20
#
def multiply_nums(inputnumber):
    newlist = [int(i) for i in inputnumber.split(", ")]
    product = 1
    for j in newlist:
        product = product * j
    return print(product)


inputnum = input("Enter the numbers to multiply seperated by commas: ")
multiply_nums(inputnum)
```

```python
# Question 3
# Create a function that squares every digit of a number.
# Examples
# square_digits(9119) ➞ 811181
# square_digits(2483) ➞ 416649
# square_digits(3212) ➞ 9414
# Notes
# The function receives an integer and must return an integer.
#
def square_digits(inputnumber, newnumlist):
    remnumber = inputnumber % 10
    newnumber = int(inputnumber / 10)
    if newnumber != 0:
        newnumlist.append(remnumber)
        square_digits(newnumber, newnumlist)
    else:
        newnumlist.append(inputnumber)
        newnumlist.reverse()
        squaredstring = ""
        for i in newnumlist:
            squaredstring = squaredstring + str(i ** 2)
        return print(squaredstring)


inputnum = int(input("Enter the number whose individual numbers have to be squared: "))
square_digits(inputnum, [])
```

```python
# Question 4
# Create a function that sorts a list and removes all duplicate items from it.
# Examples
# setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5]
# setify([4, 4, 4, 4]) ➞ [4]
# setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15]
# setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]
#
def setify(inputnumber):
    inputnumber = inputnumber.strip("[ ]")
    newlist = [int(i) for i in inputnumber.split(",")]
    newlist2 = sorted(list(set(newlist)))


    return print(newlist2)


inputnum = input("Enter a list of numbers to sort and remove duplicates: ")
setify(inputnum)
```

```python
# Question 5
# Create a function that returns the mean of all digits.
# Examples
# mean(42) ➞ 3
# mean(12345) ➞ 3
# mean(666) ➞ 6
# Notes
#  The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in
# 512 is (5+1+2)/3(number of digits) = 8/3=2).
#  The mean will always be an integer.
def mean(inputnumber, newnumlist):
    remnumber = inputnumber % 10
    newnumber = int(inputnumber / 10)


    if newnumber != 0:
        newnumlist.append(remnumber)
        mean(newnumber, newnumlist)


    else:
        newnumlist.append(inputnumber)
        length = int(len(newnumlist))
        sum = 0
        for i in newnumlist:
            sum = sum + i
        return print(int(sum / length))


inputnum = int(input("Enter the number to check mean of individual numbers: "))
mean(inputnum, [])
```
