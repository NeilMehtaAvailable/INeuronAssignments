```python
# Question2
# Create the function that takes a list of dictionaries and returns the sum of people's budgets.
# Examples
# get_budgets([
# { "name": "John", "age": 21, "budget": 23000 },
# { "name": "Steve", "age": 32, "budget": 40000 },
# { "name": "Martin", "age": 16, "budget": 2700 }
# ]) ➞ 65700
# get_budgets([
# { "name": "John", "age": 21, "budget": 29000 },
# { "name": "Steve", "age": 32, "budget": 32000 },
# { "name": "Martin", "age": 16, "budget": 1600 }
# ]) ➞ 62600
#
def get_budgets(stringlist):
    newstring = ""
    counter = 0
    newstring2 = ""
    listdict2  = []
    listdict3 = []
    listdict4 = []
    listdict5 = []
    budgetlist = 0
    for i in stringlist:
        if i == "[":
            newstring = newstring
        elif i == "]":
            newstring = newstring
        elif i != "}":
            newstring = newstring + i
        else:
            newstring = newstring + "}}"

    listdict = newstring.split("},")
    newstring2 = listdict[-1][:-1]
    listdict[-1] = newstring2

    for j in range(0, int(len(listdict))):
        listdict2.append(listdict[j].strip(" {}"))

    for k in range(0, int(len(listdict2))):
        listdict3.append(listdict2[k].split(","))

    for l in range(0, int(len(listdict3))):
        for m in range(0, int(len(listdict3[l]))):
            if m == int(len(listdict3[l])) - 1:
                listdict4.append(listdict3[l][m])

    for  n in listdict4:
        listdict5.append(int(n.replace(' "budget":',"")))

    budgetlist = sum(listdict5)


    return print(budgetlist)


inputlist = input("Enter a list of dictionaries for budget calculation: ")
get_budgets(inputlist)
```

```python
# Question3
# Create a function that takes a string and returns a string with its letters in alphabetical order.
# Examples
# alphabet_soup("hello") ➞ "ehllo"
# alphabet_soup("edabit") ➞ "abdeit"
# alphabet_soup("hacker") ➞ "acehkr"
# alphabet_soup("geek") ➞ "eegk"
# alphabet_soup("javascript") ➞ "aacijprstv"
#
def aplhabet_soup(inputstring):
    outputstr = ""
    for i in sorted(inputstring):
        outputstr = outputstr + i
    return print(outputstr)


inputuser = input("Enter a word: ")
aplhabet_soup(inputuser)
```

```python
# Question4
# Suppose that you invest $10,000 for 10 years at an interest rate of 6% compounded monthly.
# What will be the value of your investment at the end of the 10 year period?
# Create a function that accepts the principal p, the term in years t, the interest rate r, and the
# number of compounding periods per year n. The function returns the value at the end of term
# rounded to the nearest cent.
# For the example above:
# compound_interest(10000, 10, 0.06, 12) ➞ 18193.97
# Note that the interest rate is given as a decimal and n=12 because with monthly compounding
# there are 12 periods per year. Compounding can also be done annually, quarterly, weekly, or
# daily.
# Examples
# compound_interest(100, 1, 0.05, 1) ➞ 105.0
# compound_interest(3500, 15, 0.1, 4) ➞ 15399.26
# compound_interest(100000, 20, 0.15, 365) ➞ 2007316.26
#
def compound_interest(p, t, r, n):
    if p == 0 or t == 0 or r == 0 or n == 0:
        return print("Enter a non-zero value")
    return print(round((p * (1 + (r/n)) ** (t*n)), 2))

principle = input("Enter the principle amount: ")
years = input("Enter the tenure in years for the amount: ")
ratedivby100 =  input("Enter the rate of interest divided by hundred: ")
comppriod = input("Enter the number of times the principle gets compouded a year: ")
compound_interest(int(principle), float(years), float(ratedivby100), int(comppriod))
```

```python
# Question5
# Write a function that takes a list of elements and returns only the integers.
# Examples
# return_only_integer([9, 2, "space", "car", "lion", 16]) ➞ [9, 2, 16]
# return_only_integer(["hello", 81, "basketball", 123, "fox"]) ➞ [81, 123]
# return_only_integer([10, "121", 56, 20, "car", 3, "lion"]) ➞ [10, 56, 20,
# 3]
# return_only_integer(["String", True, 3.3, 1]) ➞ [1]
def return_only_integer(listinput):
    listinput1 = listinput.replace("[", "").replace("]", "")
    listinput2 = listinput1.split(",")
    collectorlist = []
    for i in listinput2:
        try:
            collectorlist.append(int(i))
        except TypeError:
            pass
        except ValueError:
            pass

    return print(collectorlist)


iputlist = input("Enter a list of integer and string of mixed data types: ")
return_only_integer(iputlist)
```
