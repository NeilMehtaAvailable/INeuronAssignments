```python
##Write a Python program to check if the given number is Happy Number?
def NumberSquaring(n):
    squarednum = 0
    while (n):
        squarednum = squarednum + (n % 10) * (n % 10)
        n = int(n / 10)
    return squarednum


def HappyNumTest(x):
    testnum1 = x
    testnum2 = x
    while True:
        testnum1 = NumberSquaring(testnum1)
        testnum2 = NumberSquaring(NumberSquaring(testnum2))
        if testnum1 == testnum2:
            break
    return testnum1 == 1


testnum = int(input("Enter INT to check whether the number is Happy number or not: "))


if (HappyNumTest(testnum)):
    print("The given number is a Happy number")
else:
    print("The given number is not a Happy number")
```

```python
##Write a Python program to print all happy numbers between 1 and 100?
def NumberSquaring(n):
    squarednum = 0
    while (n):
        squarednum = squarednum + (n % 10) * (n % 10)
        n = int(n / 10)
    return squarednum


def HappyNumTest(x):
    testnum1 = x
    testnum2 = x
    while True:
        testnum1 = NumberSquaring(testnum1)
        testnum2 = NumberSquaring(NumberSquaring(testnum2))
        if testnum1 == testnum2:
            break
    return testnum1 == 1


testintupperlimit = int(input("Enter the upper limit you want to display all the Happy numbers:"))
testnumlist = []
for j in range(1, testintupperlimit + 1):
    testnumlist.append(j)


for i in testnumlist:
    if (HappyNumTest(i)):
        print("This number is a Happy number: ", i)
```

```python
##Write a Python program to determine whether the given number is a Harshad Number?
testintnum = int(input("Enter INT to check whether the number is Harshad number or not: "))
teststrnum = str(testintnum)
lengthstrnum = len(teststrnum)
testintnum2 = testintnum
indinumlist = []


while True:
    remindernum = testintnum2 % 10
    testintnum2 = int(testintnum2 / 10)
    indinumlist.append(remindernum)
    if len(indinumlist) == lengthstrnum:
        break

if (testintnum % sum(indinumlist)) == 0:
    print("The number is a Harshad number")
else:
    print("The number is not a Harshad number")
```
