```python
# Question 1:
# Define a class with a generator which can iterate the numbers, which are divisible by
# 7, between a given range 0 and n.
#
class divisibilitytest:
    def __init__(mypointer, upperlimit):
        mypointer.accumulatorr = []
        mypointer.upperlimit = upperlimit


    def divisibilitytest_func(mypointer):
        for i in range(0, int(mypointer.upperlimit) + 1):
            if i % 7 == 0:
                mypointer.accumulatorr.append(i)
        mypointer.mylist = mypointer.accumulatorr


upperlimit = input("Enter the upper limit upto which you want to see 7's table:")
resultlist = divisibilitytest(upperlimit)
resultlist.divisibilitytest_func()
print("The list is:",resultlist.mylist)
```

```python
# Question 2:
# Write a program to compute the frequency of the words from the input. The output
# should output after sorting the key alphanumerically.
# Suppose the following input is supplied to the program:
# New to Python or choosing between Python 2 and Python 3? Read Python 2 or
# Python 3.
# Then, the output should be:
# 2:2
# 3.:1
# 3?:1
# New:1
# Python:5
# Read:1
# and:1
# between:1
# choosing:1
# or:2
# to:1
#
class wordcounter:
    def __init__(mypointer, myinputstring):
        mypointer.inputstring = myinputstring
        mypointer.inputwordlist = []
        mypointer.inputwordlistsorted = []
        mypointer.inputwordlistsortedunique = []
        mypointer.elementcounterlist = []
        mypointer.outputstr = ""


    def string2listconverter(mypointer):
        mypointer.inputwordlist = mypointer.inputstring.split(' ')


    def listsorter(mypointer):
        mypointer.inputwordlistsorted = sorted(mypointer.inputwordlist)


    def uniqueelementorderedlist(mypointer):
        mypointer.inputwordlistsortedunique = []
        for l in mypointer.inputwordlistsorted:
            if l not in mypointer.inputwordlistsortedunique:
                mypointer.inputwordlistsortedunique.append(l)


    def elementcounter(mypointer):
        for i in mypointer.inputwordlistsortedunique:
            mypointer.elementcounterlist.append(mypointer.inputwordlistsorted.count(i))


    def list2stringconverter(mypointer):
        for j in range(0, int(len(mypointer.inputwordlistsortedunique))):
            mypointer.outputstr = mypointer.outputstr + mypointer.inputwordlistsortedunique[j] + ":" + str(mypointer.elementcounterlist[j]) + "\n"


inputstring = input("Enter a sentence to check each word's frequency occurance: ")
resultclass = wordcounter(inputstring)
resultclass.string2listconverter()
resultclass.listsorter()
resultclass.uniqueelementorderedlist()
resultclass.elementcounter()
resultclass.list2stringconverter()
print("The alphanumericaly sorted and frequency of words are:", resultclass.outputstr)
```

```python
# Question 3:
#
# Define a class Person and its two child classes: Male and Female. All classes have a
# method "getGender" which can print "Male" for Male class and "Female" for Female
# class.
#
class Person:
    def __init__(mypointer, getGender):
        mypointer.gender = getGender


class Male(Person):
    def malegender(mypointer):
        if mypointer.gender.lower() == 'male':
            print("You entered Male")


class Female(Person):
    def femalegender(mypointer):
        if mypointer.gender.lower() == 'female':
            print("You entered Female")


class Unidentified(Person):
    def noneoftheabove(mypointer):
        if mypointer.gender.lower() != 'female' or mypointer.gender.lower() != 'male':
            print("Either enter male or female")


inputstring = input("Enter either male or female: ")


resultclass1 = Male(inputstring)
resultclass1.malegender()


resultclass2 = Female(inputstring)
resultclass2.femalegender()


resultclass3 = Unidentified(inputstring)
resultclass3.noneoftheabove()
```

```python
# Question 4:
# Please write a program to generate all sentences where subject is in ["I", "You"] and
# verb is in ["Play", "Love"] and the object is in ["Hockey","Football"].
#
import numpy as np
class subjectclass:
    def __init__(self, subjectlist, verblist, objectlist):
        self.subject = subjectlist
        self.verb = verblist
        self.object = objectlist
        self.wordmatrix = np.empty((3,2), dtype = "<U15")
        self.sentencebuilt = ''


    def sentencebuilderfunc1(self):
        for i in range(0, int(len(self.subject))):
            self.wordmatrix[0,i] = self.subject[i]


class verbclass(subjectclass):
    def sentencebuilderfunc2(self):
        for j in range(0, int(len(self.verb))):
            self.wordmatrix[1,j] = self.verb[j]


class objectclass(verbclass):
    def sentencebuilderfunc3(self):
        for k in range(0, int(len(self.object))):
            self.wordmatrix[2,k] = self.object[k]


class sentencebuilderclass(objectclass):
    def sentencebuilderfunc(self):
        for l in self.wordmatrix[0]:
            for m in self.wordmatrix[1]:
                for n in self.wordmatrix[2]:
                    self.sentencebuilt = self.sentencebuilt + l + ' ' + m + ' ' + n + "\n"


inputsubject = ["I", "You"]
inputverb = ["Play", "Love"]
inputobject = ["Hockey", "Football"]


resultsentencebuilderclass = sentencebuilderclass(inputsubject, inputverb, inputobject)
resultsentencebuilderclass.sentencebuilderfunc1()
resultsentencebuilderclass.sentencebuilderfunc2()
resultsentencebuilderclass.sentencebuilderfunc3()
resultsentencebuilderclass.sentencebuilderfunc()


print("All the combination  of sentences are:", resultsentencebuilderclass.sentencebuilt)
```

```python
# Question 5:
# Please write a program to compress and decompress the string "hello world!hello
# world!hello world!hello world!".
#
class compression:
    def __init__(self, inputstring):
        self.inputstring = inputstring
        self.compstresult = ""
        self.compcounter = 0
        self.compinitvalue = inputstring[0]
        self.decompresult = ""
        self.decompcounter = 0
        self.decompinitvalue = ""


    def compalgo(self):
        for i in self.inputstring + " ":
            if i == self.compinitvalue:
                self.compcounter += 1
            else:
                self.compstresult = self.compstresult + str(self.compcounter) + self.compinitvalue
                self.compinitvalue = i
                self.compcounter = 1


class decompression(compression):
    def decompalgo(self):
        self.decompinitvalue = self.compstresult[0]
        for j in self.compstresult + " ":
            if j.isnumeric() is False:
                self.decompinitvalue = j
                self.decompresult = self.decompresult + self.decompcounter * self.decompinitvalue
            else:
                self.decompcounter = int(j)


inputsentence = "hello world!hello world!hello world!hello world!"
decompressionresult = decompression(inputsentence)
decompressionresult.compalgo()
decompressionresult.decompalgo()
print("The resultant compression of input sentence is:", decompressionresult.compstresult)
print("The resultant decompression of input sentence is:", decompressionresult.decompresult)
```

```python
# Question 6:
# Please write a binary search function which searches an item in a sorted list. The
# function should return the index of element to be searched in the list.
#
#This answer is only for integer or floating inputs. Not any other data type inputs
class sortingclass:
    def __init__(self, orignallist, targetcharacter):
        self.orignallist = orignallist
        self.sortedlist = []
        self.lowestindex = 0
        self.upperindex = int(len(self.orignallist)) - 1
        self.midindex = 0
        self.targetchar = targetcharacter
        self.consideredchar = 0
        self.targetindex = 0


    def sortingfunc(self):
        self.sortedlist = sorted(self.orignallist)


class searchclass(sortingclass):
    def searchingfunc(self):
        while True:
            self.midindex = int((self.lowestindex + self.upperindex) / 2)
            self.consideredchar = self.sortedlist[self.midindex]
            if self.targetchar == self.consideredchar or self.targetchar == self.sortedlist[-1]:
                self.targetindex = int(self.orignallist.index(self.targetchar))
                break
            elif self.targetchar > self.consideredchar:
                self.lowestindex = int(self.sortedlist.index(self.consideredchar))
            else:
                self.upperindex = int(self.sortedlist.index(self.consideredchar))


InputList = [1,3,2,5,4,7,8,6]
TargetElement = 5
Result = searchclass(InputList, TargetElement)
Result.sortingfunc()
Result.searchingfunc()
print("The Index of target element in the list is:", Result.targetindex)
```
