```python
# Question1
# Create a function that takes a list of strings and integers, and filters out the list so that it
# returns a list of integers only.
# Examples
# filter_list([1, 2, 3, "a", "b", 4]) ➞ [1, 2, 3, 4]
# filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) ➞ [0, 1729]
# filter_list(["Nothing", "here"]) ➞ []
#
def filter_list(inputlist):
    usablelist1 = inputlist.split(",")
    usablelist2 = []
    usablelist3 = []
    finallist = []
    for i in range(0, int(len(usablelist1))):
        for j in usablelist1[i]:
            usablelist2.append(usablelist1[i].replace(" ","").replace("[", "").replace("]", ""))
            break


    for k in usablelist2:
        if k.isdigit():
            usablelist3.append(int(k))


    return print(usablelist3)


inputlist = input("Enter the list of strings and integers: ")
filter_list(inputlist)
```

```python
# Question2
# Given a list of numbers, create a function which returns the list but with each element's
# index in the list added to itself. This means you add 0 to the number at index 0, add 1 to the
# number at index 1, etc...
# Examples
# add_indexes([0, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4]
# add_indexes([1, 2, 3, 4, 5]) ➞ [1, 3, 5, 7, 9]
# add_indexes([5, 4, 3, 2, 1]) ➞ [5, 5, 5, 5, 5]
#
def add_indexes(numberlist):
    usablelist1 = numberlist.split(",")
    usablelist2 = []
    usablelist3 = []
    indexlist = []
    summedlist = []


    for i in range(0, int(len(usablelist1))):
        for j in usablelist1[i]:
            usablelist2.append(usablelist1[i].replace(" ","").replace("[", "").replace("]", ""))
            break


    for j in usablelist2:
        if j.isdigit():
            usablelist3.append(int(j))


    for k in range(0, int(len(usablelist3))):
        indexlist.append(k)


    for l in range(0, int(len(usablelist3))):
        summedlist.append(usablelist3[l] + indexlist[l])


    return print(summedlist)


inputlist = input("Enter the list of integers: ")
add_indexes(inputlist)
```

```python
# Question4
# This Triangular Number Sequence is generated from a pattern of dots that form a triangle.
# The first 5 numbers of the sequence, or dots, are:
# 1, 3, 6, 10, 15
# This means that the first triangle has just one dot, the second one has three dots, the third one
# has 6 dots and so on.
# Write a function that gives the number of dots with its corresponding triangle number of the
# sequence.
# Examples
# triangle(1) ➞ 1
# triangle(6) ➞ 21
# triangle(215) ➞ 23220
#
def triangle(n):
    return print(int(int(n) * (int(n) + 1) / 2))

inputrows = input("Enter the number of rows: ")
triangle(inputrows)
```

```python
# Question5
# Create a function that takes a list of numbers between 1 and 10 (excluding one number) and
# returns the missing number.
# Examples
# missing_num([1, 2, 3, 4, 6, 7, 8, 9, 10]) ➞ 5
# missing_num([7, 2, 3, 6, 5, 9, 1, 4, 8]) ➞ 10
# missing_num([10, 5, 1, 2, 4, 6, 8, 3, 9]) ➞ 7
def missing_num(inputlist):
    usablelist1 = inputlist.split(",")
    usablelist2 = []
    usablelist3 = []


    for i in range(0, int(len(usablelist1))):
        for j in usablelist1[i]:
            usablelist2.append(usablelist1[i].replace(" ", "").replace("[", "").replace("]", ""))
            break

    for j in usablelist2:
        if j.isdigit():
            usablelist3.append(int(j))


    for k in range(1, 11):
        if k not in usablelist3:
            return print(k)


inputlist = input("Enter the list of integers: ")
missing_num(inputlist)
```
