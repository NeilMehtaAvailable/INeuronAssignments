```python
# Question 1
# Create a function that takes a list of non-negative integers and strings and return a new list
# without the strings.
# Examples
# filter_list([1, 2, "a", "b"]) ➞ [1, 2]
# filter_list([1, "a", "b", 0, 15]) ➞ [1, 0, 15]
# filter_list([1, 2, "aasf", "1", "123", 123]) ➞ [1, 2, 123]
#
def outputnumlist(listinput):
    listinput1 = listinput.replace("[", "").replace("]", "")
    listinput2 = listinput1.split(",")
    collectorlist = []
    for i in listinput2:
        try:
            collectorlist.append(int(i))
        except TypeError:
            pass
        except ValueError:
            pass

    return print(collectorlist)


iputlist = input("Enter a list of integer and string of mixed data types: ")
outputnumlist(iputlist)
```

```python
# Question 2
# The "Reverser" takes a string as input and returns that string in reverse order, with the
# opposite case.
# Examples
# reverse("Hello World") ➞ "DLROw OLLEh"
# reverse("ReVeRsE") ➞ "eSrEvEr"
# reverse("Radar") ➞ "RADAr"
#
def Reverser(inputword):
    inputlist = list(inputword)
    collectorstr = ""
    collectorlist = []
    reversedcollecterstr = ""
    for i in inputlist:
        if i == " ":
            collectorstr = collectorstr + " "
        elif i.isalpha():
            if i.islower():
                collectorstr = collectorstr + i.upper()
            else:
                collectorstr = collectorstr + i.lower()


    collectorlist = collectorstr.split(" ")
    for j in collectorlist:
        reversedcollecterstr = reversedcollecterstr + j[::-1] + " "


    return print(reversedcollecterstr)


inputstring = input("Enter the word for swapcase and reverse sequence: ")
Reverser(inputstring)
```

```python
# Question 3
# You can assign variables from lists like this:
# lst = [1, 2, 3, 4, 5, 6]
# first = lst[0]
# middle = lst[1:-1]
# last = lst[-1]
# print(first) ➞ outputs 1
# print(middle) ➞ outputs [2, 3, 4, 5]
# print(last) ➞ outputs 6
# With Python 3, you can assign variables from lists in a much more succinct way. Create
# variables first, middle and last from the given list using destructuring assignment
# (check the Resources tab for some examples), where:
# first ➞ 1
# middle ➞ [2, 3, 4, 5]
# last ➞ 6
#
# Your task is to unpack the list writeyourcodehere into three variables, being first,
# middle, and last, with middle being everything in between the first and last element. Then
# print all three variables.
#
lst = [1, 2, 3, 4, 5, 6]
first = lst[0]
middle = lst[1:-1]
last = lst[-1]
print(first, middle, last)


def mmethod1(lst):
    first, _, _, _, _, last = lst
    middle = [lst[i] for i in range(1, int(len(lst)) - 1)]
    return print(first, middle, last)


def method2(lst):
    first, *last = lst
    middle = [lst[i] for i in range(1, int(len(lst)) - 1)]
    last = last[-1]
    return print(first, middle, last)


mmethod1(lst)
method2(lst)
```

```python
# Question 4
# Write a function that calculates the factorial of a number recursively.
# Examples
# factorial(5) ➞ 120
# factorial(3) ➞ 6
# factorial(1) ➞ 1
# factorial(0) ➞ 1
#
def factorialfunc(x, previousnum):
    if x > 1:
        previousnum = previousnum * x
        y = x - 1
        factorialfunc(y, previousnum)
    elif x == 1:
        return print(previousnum)
    else:
        return print("Positive integers only")



factorialinput = int(input("Enter the number whose factorial you want to find: "))
factorialfunc(factorialinput, 1)
```

```python
# Question 5
# Write a function that moves all elements of one type to the end of the list.
# Examples
# move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]
# # Move all the 1s to the end of the array.
# move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]
# move_to_end(["a", "a", "a", "b"], "a") ➞ ["b", "a", "a", "a"]
def sorterlist(listinput):
    usablelist1 = []
    specialcharindex = 0
    usablelist2 = []
    usablelist3 = []
    finallist1 = []
    finallist = []
    counter = 0
    for i in range(1, int(len(listinput))):
        if listinput[i] != "," and listinput[i] != " ," and listinput[i] != " " and listinput[i] != '"' and listinput[i] != "'":
            usablelist1.append(listinput[i])


    specialcharindex = int(usablelist1.index("]"))
    usablelist1.pop(specialcharindex)


    for j in range(0, specialcharindex):
        usablelist2.append(usablelist1[j])


    usablelist3 = [usablelist2]
    usablelist3.append(usablelist1[-1])


    for k in usablelist3[0]:
        if k == usablelist3[-1]:
            counter += 1
        else:
            finallist1.append(k)


    if counter >=  1:
        for l in range(0, counter + 1):
            finallist1.append(usablelist3[-1])
    else:
        finallist1.append(finallist1)


    for m in finallist1:
        if m.isdigit():
            m = int(m)
            finallist.append(m)
        else:
            finallist.append(m)


    return print(finallist)


iputlist = input("Enter a list of integers or strings in format [x1, x2, x3.....], x2: ")
sorterlist(iputlist)
```
