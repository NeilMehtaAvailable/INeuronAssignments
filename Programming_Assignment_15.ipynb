```python
#Question 1:
#Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console.
#Example:
#If the following n is given as input to the program:
#100
#Then, the output of the program should be:
#0,35,70
#

def commondivisibility(maxnum):
    upperlimit = int(maxnum)
    lowerlimit = 0
    collectorlist5 = []
    collectorlist7 = []
    commonelements = []
    collectorset5 = {}
    collectorset7 = {}
    k = 0
    collectorstring = ""


    for i in range(lowerlimit, upperlimit + 1):
        if i % 5 == 0:
            collectorlist5.append(i)


    for j in range(lowerlimit, upperlimit + 1):
        if j % 7 == 0:
            collectorlist7.append(j)


    collectorset5 = set(collectorlist5)
    collectorset7 = set(collectorlist7)
    commonelements = list(collectorset5 & collectorset7)
    commonelements = sorted(commonelements)
    while True:
        if int(len(commonelements)) == 1:
            collectorstring = "The only common numer is 0. Re-execute and increase the limit"
            yield collectorstring
            break
        elif k > int(len(commonelements)):
            break
        elif k < int(len(commonelements)):
            collectorstring = str(commonelements[k])
            yield collectorstring
            k += 1
            if k == int(len(commonelements)):
                break


inputnum = int(input("Enter the maximum nu. upto which you want to see common multiples of 5 and 7: "))
resultstring = commondivisibility(inputnum)
print("The common multiples are:")
cumilativestring = ""
for l in resultstring:
    cumilativestring = cumilativestring + str(l) + ","
print(cumilativestring,"\b\b")
```

```python
#Question 2:
#Please write a program using generator to print the even numbers between 0 and n in comma separated form while n is input by console.
#Example:
#If the following n is given as input to the program:
#10
#Then, the output of the program should be:
#0,2,4,6,8,10
#
def evennumbers(upperlimit):
    lowerlimit = 0
    evennumstring = ""
    k = -1
    while True:
        k += 1
        if k % 2 == 0 and k <= upperlimit:
            evennumstring = k
            yield evennumstring
        elif k > upperlimit:
            break


enternum = int(input("Enter the upper limit upto which  you want to see all even numbers: "))
outputfunc = evennumbers(enternum)
cumilativestring = ""
for l in outputfunc:
    cumilativestring = cumilativestring + str(l) + ","
print(cumilativestring,"\b\b")
```

```python
#Question 3:
#The Fibonacci Sequence is computed based on the following formula:
#f(n)=0 if n=0
#f(n)=1 if n=1
#f(n)=f(n-1)+f(n-2) if n>1
#Please write a program using list comprehension to print the Fibonacci Sequence in comma separated form with a given n input by console.
#Example:
#If the following n is given as input to the program:
#7
#
#Then, the output of the program should be:
#0,1,1,2,3,5,8,13
#
def fibbo(numofterms):
    outputnum = [0, 1]
    [outputnum.append(outputnum[counter - 1] + outputnum[counter - 2]) for counter in range(2, numofterms)]
    yield outputnum


inputnum = int(input(("Enter the number of terms you want to see as Fibo nums: ")))
result = fibbo(inputnum)
cumilativestring = ""
for l in result:
    cumilativestring = cumilativestring + str(l) + ","
print(cumilativestring.replace("[", "").replace("]", ""),"\b\b")
```

```python
#Question 4:
#Assuming that we have some email addresses in the "username@companyname.com" format, please write program to print the user name of a given email address. Both user names and company names are composed of letters only.
#Example:
#If the following email address is given as input to the program:
#john@google.com
#Then, the output of the program should be:
#john
#
def partitionmethod(emailid):
    username = emailid.partition("@")[0]
    return username


inputusername = input("Enter the emailid of the person whose username you want to know: ")
usernameouput = partitionmethod(inputusername)
print("Username is:", usernameouput)
```
