```python
# Question 1:
# Write a program that calculates and prints the value according to the given formula:
# Q = Square root of [(2 * C * D)/H]
# Following are the fixed values of C and H:
# C is 50. H is 30.
# D is the variable whose values should be input to your program in a comma-separated
# sequence.
# Example
# Let us assume the following comma separated input sequence is given to the program:
# 100,150,180
# The output of the program should be:
# 18,22,24
import math
C = 50
H = 30
E = []
Q = []
E.append(list(input("Enter input values seperated by commas:").split(',')))
D = E[0]

for i in range(0, int(len(D))):
    D[i] = int(D[i])

for i in D:
    Q.append(int(math.sqrt((2 * C * i)/H)))


print("The output of formula is: ", Q)
```

```python
# Question 2:
# Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
# element value in the i-th row and j-th column of the array should be i*j.
# Note: i=0,1.., X-1; j=0,1,Â¡Y-1.
# Example
# Suppose the following inputs are given to the program:
# 3,5
# Then, the output of the program should be:
# [[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]]
#
import numpy
sizelist = []
sizelist.append(list(input("Enter the dimensions seperated by commas:").split(',')))
sizelist = sizelist[0]
mymatrix = numpy.zeros([int(sizelist[0]), int(sizelist[1])], dtype= int)
for i in range(0, list(mymatrix.shape)[0]):
    for j in range(0, list(mymatrix.shape)[1]):
        mymatrix[i][j] = i*j
print("The resultant matrix is: ", mymatrix)
```

```python
# Question 3:
# Write a program that accepts a comma separated sequence of words as input and prints the
# words in a comma-separated sequence after sorting them alphabetically.
# Suppose the following input is supplied to the program:
# without,hello,bag,world
# Then, the output should be:
# bag,hello,without,world
#
inputwordlist = [(input("Enter input values seperated by commas and without spaces:").split(','))]
inputwordlist = sorted(inputwordlist[0])
inputstr = ""
for i in inputwordlist:
    inputstr = inputstr + i + " "
print("The sorted string is:", inputstr)
```

```python
# Question 4:
# Write a program that accepts a sequence of whitespace separated words as input and prints
# the words after removing all duplicate words and sorting them alphanumerically.
# Suppose the following input is supplied to the program:
# hello world and practice makes perfect and hello world again
# Then, the output should be:
# again and hello makes perfect practice world
#
inputwordlist = [(input("Enter input values seperated by spaces:").split(' '))][0]
inputwordset = set(inputwordlist)
inputwordset = sorted(inputwordset)
inputstr = ""
for i in inputwordset:
    inputstr = inputstr + i + " "
print("The sorted string is:", inputstr)
```

```python
# Question 5:
# Write a program that accepts a sentence and calculate the number of letters and digits.
# Suppose the following input is supplied to the program:
# hello world! 123
# Then, the output should be:
# LETTERS 10
#
# DIGITS 3
#
inputwordlist = [(input("Enter input values:"))][0]
countalpha =  0
countnum = 0
for i in inputwordlist:
    if i.isalpha():
        countalpha += 1
    elif i.isdigit():
        countnum += 1

print("The number of aplhabets in input are", countalpha,"and the number of numbers are", countnum)
```

```python
# Question 6:
# A website requires the users to input username and password to register. Write a program to
# check the validity of password input by users.
# Following are the criteria for checking the password:
# 1. At least 1 letter between [a-z]
# 2. At least 1 number between [0-9]
# 1. At least 1 letter between [A-Z]
# 3. At least 1 character from [$#@]
# 4. Minimum length of transaction password: 6
# 5. Maximum length of transaction password: 12
# Your program should accept a sequence of comma separated passwords and will check them
# according to the above criteria. Passwords that match the criteria are to be printed, each
# separated by a comma.
# Example
# If the following passwords are given as input to the program:
# ABd1234@1,a F1#,2w3E*,2We3345
# Then, the output of the program should be:
# ABd1234@1
inputwordlist = [(input("Enter different passwords seperated by commas:").split(','))][0]
goodpasswordlist =[]


for passwordx in inputwordlist:
    listpassorfail = [0,0,0,0,0]
    for characterx in passwordx:
        if ord(characterx) == 35 or ord(characterx) == 36 or ord(characterx) == 64:
            listpassorfail[0] = 1
        if ord(characterx) >= 48 and ord(characterx) <= 57:
            listpassorfail[1] = 1
        if ord(characterx) >= 65 and ord(characterx) <= 90:
            listpassorfail[2] = 1
        if ord(characterx) >= 97 and ord(characterx) <= 122:
            listpassorfail[3] = 1
    if int(len((passwordx))) >= 6 and int(len((passwordx))) <= 12:
        listpassorfail[4] = 1
    if sum(listpassorfail) == 5:
        goodpasswordlist.append(passwordx)


goodpasswordstring = ""
k = int(len(goodpasswordlist))
for i in range(0, int(len(goodpasswordlist))):
    if i < k - 1:
        goodpasswordstring = goodpasswordstring + goodpasswordlist[i] + ", "
    else:
        goodpasswordstring = goodpasswordstring + goodpasswordlist[i]


if int(len(goodpasswordlist)) == 0:
    print("There are no good passwords")
else:
    print("Good password(s) is/are:", goodpasswordstring)
```
